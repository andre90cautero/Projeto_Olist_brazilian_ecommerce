## Tabela Orders
### Os pedidos são entregues antes/na data/ depois da data ?

```python
df_orders.head()
```

```python
df_orders.info()
```

**para continuar a analise precisamos converter as colunas de datas de object para timestamp

```python
df_orders['order_purchase_timestamp']=pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['order_approved_at']=pd.to_datetime(df_orders['order_approved_at'])
df_orders['order_delivered_carrier_date']=pd.to_datetime(df_orders['order_delivered_carrier_date'])
df_orders['order_delivered_customer_date']=pd.to_datetime(df_orders['order_delivered_customer_date'])
df_orders['order_estimated_delivery_date']=pd.to_datetime(df_orders['order_estimated_delivery_date'])
```

### calculando os dias entre a entrega e a previsao de entrega

```python
# diferença entre data de entrega e pedido
df_orders['delivery_days'] = df_orders['order_delivered_customer_date']-df_orders['order_purchase_timestamp']
# diferença entre data de entrega e pedido
df_orders['delivery_days_calc'] = df_orders['order_estimated_delivery_date']-df_orders['order_purchase_timestamp']
```

```python
entrega = df_orders[['delivery_days','delivery_days_calc']]
entrega.columns = [['dias_entrega','dias_entrega_previsionado']]
```

```python
entrega.head()
```

### Criando as classes entregue antes / entregue no dia / entregue depois

```python
### criando condiçoes para as classes

condicoes = [ 
(df_orders['delivery_days'] < df_orders['delivery_days_calc']),
(df_orders['delivery_days'] > df_orders['delivery_days_calc']),
(df_orders['delivery_days'] == df_orders['delivery_days_calc'])  
]
opcoes = ['entregue antes', 'entregue depois', 'entregue no dia']
```

```python
entrega['status'] = np.select(condicoes, opcoes)
```

```python
entrega.head()
```

```python
total_status = entrega['status'].value_counts().sum()
df_status = entrega['status'].value_counts().to_frame().reset_index()
df_status.columns = [['status','contagem']]
df_status['%'] = (df_status['contagem']/total_status*100).round(2) 
```

```python
df_status
```

## Payments

- Para o tipo de pagamento cartão de crédito, existe relação do numero de parcelas escolhida com o valor ? 

Correlação positiva: quando duas variáveis que possuem correlação crescem ou decrescem juntas, ou seja, que possuem uma relação direta;
<br>
Correlação negativa: quando duas variáveis que possuem correlação mas quando uma variável cresce a outra decresce, ou vice-versa;
<br>
Não ter correlação: quando o crescimento ou decrescimento de uma variável não tem efeito sobre outra variável.
Como podemos interpretar os valores que ρ pode assumir?
<br>
<li>ρ = 0,9 a 1 (positivo ou negativo): correlação muito forte;</li>
<li>ρ = 0,7 a 09 (positivo ou negativo): correlação forte;</li>
<li>ρ = 0,5 a 0,7 (positivo ou negativo): correlação moderada;</li>
<li>ρ = 0,3 a 0,5 (positivo ou negativo): correlação fraca;</li>
<li>ρ = 0 a 0,3 (positivo ou negativo): não possui correlação.</li>
